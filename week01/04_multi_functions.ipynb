## 스칼라와 벡터
- 스칼라 : 양만으로 표현할 수 있는 물리량
  - 수직선상에서의 점
  - 온도, 속력, 부피, 넓이
  - 숫자 하나
- 벡터 
  - 양과 방향으로 표현할 수 있는 물리량
  - 공간의 두 점 또는 원점과 한 점으로 정의한다
  - 힘과 속도
  - 여러 개의 숫자
- 텐서
  - 벡터로도 표현하기 곤란한 차원 높은 물리량
  - 0차원부터 n차원까지의 존재하는 물리량
  - 대개 데이터분석에서는 다차원 배열이라고 생각하면 좋다


## 벡터 표기

## 다변수 벡터함수
- 입출력이 모두 볼드체로 쓰여 있는 것이 다변수 벡터함수이다.
- 입력도 벡터 출력도 벡터 이다
- (나타난 장표에서는) 함수의 개수가 n개, 데이터가 m개 이다.
- $s(u, v) = (u, v, 1 + u^2 + v / (1 + v^2))^T$
- 나타난 장표에서는 (u, v) -> (x, y, z) 로 표현된 것인데,    
  u, v는 표현하지 못하고, x, y, z의 결괏값을 찍은 것이라고 보면 된다.

## 다변수 함수의 그래프 방식 표현
- u, v가 있다고 했을 때
- $x = 1 * u + 0 * v$
- $y = 0 * u + 0 * v$
- $z = 1 + u^2 + v/(1 + v^2)$
- 인공신경망은 사실 벡터함수임을 알 수 있다.
- 다변수 함수는 인공신경망이 그대로 모델링된 것임을 알 수 있다.
- 벡터함수가 합성이 되어 있는 것이다.

## 함수의 합성
- 정의 : 함수 f:X -> Y의 공역과 함수 g:Y->Z의 정의역이 같다고 할 때 다음과 같이 정의된 함수 g * f를 두 함수 f 와 g의 합성이라고 한다.
- 표기 : g(f(x)), g * f
- 입출력 관점 : f의 출력이 g의 입력으로 들어간다.
- 사례 : 사람-선호분야와 선호분야-적합언어가 합성된 새로운 관계(사람-적합언어)가 창출된다

## 인공신경망 또한 합성함수
- 신경망은 매우 많은 함수가 다음처럼 겹겹이 합성된 것이라 할 수 있다.
- 인공신경망은 다변수 벡터함수가 되는 것이다.
- 함수 합성시킨 다음 로지스틱 시그모이드 함수를 합성시킨다
  - 로지스틱 시그모이드 함수 : 일변수 스칼라 함수
- 사실 인공지능 역시 표현력이 큰 함수를 놓고 그 알맹이를 채워나가는 과정이다.
